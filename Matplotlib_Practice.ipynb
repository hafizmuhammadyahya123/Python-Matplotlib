```python
# bar plot
import matplotlib.pyplot as plt
import numpy as np

py_libraries = ["pandas","numpy","seaborn","scikitlearn"]
work = ["ML","NUMERICAL","GRAPH","ML"]

width = 0.2

p = np.arange(len(py_libraries))
p1 = [j + width for j in p]

plt.xlabel("ABCISSA",fontsize=14,color = 'r')
plt.ylabel("Range",fontsize=14,color = 'r')
plt.title("PYTHON LIBRARIES" , fontsize=12 , color = 'g')

plt.bar(py_libraries,work , color = 'b' , width = 0.4)
plt.bar(py_libraries,work , color = 'b' , width = 0.4)
plt.xticks(p+width/2 , py_libraries)

plt.show()
```
