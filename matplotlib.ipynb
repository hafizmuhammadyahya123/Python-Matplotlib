```python
# BAR PLOT
import matplotlib.pyplot as plt
import numpy as np

x = ["python", "c", "c++", "java"]
y = [90, 60, 50, 91]
z = [10, 20, 30, 40]


width = 0.2
p = np.arange(len(x))
p1 = [j + width for j in p]


plt.bar(p,y, width, color = "b", align = "center", edgecolor = "black", linewidth = 5, label = "popularity")
plt.bar(p1,z, width, color =  "y", edgecolor = "r", linewidth = 3, alpha = 0.8, label = "popularity1")

plt.xlabel("language",fontsize = 20)
plt.ylabel("number",fontsize = 20)
plt.title("BAR PLOT",fontsize = 25)

plt.xticks(p+width/2 , x , rotation = 10)
plt.legend()
plt.show()






```

```python
import matplotlib.pyplot as plt
import numpy as np

pop_coun = ["USA" , "CHINA" , "JAPAN" , "RUSSIA"]
year = ["2013(8.3)" , "2015(9.3)" , "1019(9.1)" , "2020(5.2)"]
addnew_bar = ["2010" , "2014" , "208" , "2019"]

width = 0.3
p = np.arange(len(pop_coun))
p1 = [j + width for j in p]


plt.bar(p , year , width , color = "y" , edgecolor = "b" ,  linewidth = 2 , alpha = 0.8 , label = "popularity") #alpha or legend() = control bar brightness
plt.bar(p1 , addnew_bar , width , color = "r" , edgecolor = "black" , linewidth = 2 , label = "popularity 1")
plt.legend()

plt.xlabel("ABCISSA" , fontsize = 15)
plt.ylabel("RANGE" , fontsize = 15 )

plt.title("POPULATION MONITORING.." , fontsize = 15)
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

mult_datatype = ['a' , "APPLE" , '5' , '5.0' , True]
ind_lst = [0 , 1 , 2 , 3 , 4]
l = [5 , 4 , 3 , 2 , 1]

width = 0.5
p = np.arange(len(mult_datatype))
p_1  = [j + width for j in p]


plt.bar(p , ind_lst , width , color = "y" , edgecolor = "b" ,  linewidth = 2 , alpha = 0.8 , label = "popularity") #label or legend() dono k saath use krna ha always
plt.bar(p_1 , l , width , color = "b" , edgecolor = "black" , linewidth = 2 , label = "popularity 1")
plt.legend()

plt.xticks(p+width/2 , mult_datatype , rotation = 2)
plt.xlabel("DOMAIN" , fontsize = 30 , color = 'blue')
plt.ylabel("CODOMAIN" , fontsize = 30 , color = 'blue')

plt.title("BAR PLOT...." , fontsize = 40 , color = "green")
plt.show()
```

```python

import matplotlib.pyplot as plt
list = [1 , 2 , 3 , 4 , 5 , 6 , 7 , 8]
list_A = [0 , 1 , 2 , 6 ,  3 , 9, 8 , 3]
colors = ['r', 'y', 'b', 'r', 'g', 'r', 'g', 'b']
sizes = [100 , 290 , 500 , 110 , 200 , 300 , 600 , 120]
plt.scatter(list , list_A , c = colors , alpha = 0.6 , s = sizes , marker="*" , edgecolor = 'black' , linewidths=2)#alpha = brightness control 0 to 1 k middle k all float num
plt.xlabel("horizontal" , fontsize = 15 , color = "blue")
plt.ylabel("vertical" , fontsize = 15  , color = "blue")
plt.title("SCATTER" , fontsize = 25 , color = "blue")
plt.show()
```

```python
import matplotlib.pyplot as plt
list = [1 , 2 , 3 , 4 , 5 , 6 , 7 , 8]
list_A = [0 , 1 , 2 , 6 ,  3 , 9, 8 , 3]
list_B = [2, 4, 6, 7, 9, 1, 3, 8]
colors = [10, 40, 50, 45, 90, 30, 20, 65]

sizes = [100 , 290 , 500 , 110 , 200 , 300 , 600 , 120]
plt.scatter(list , list_A , c = colors , cmap="viridis" , s = sizes)
plt.scatter(list , list_B , color = 'r' , s = sizes)

t = plt.colorbar()
t.set_label("COLOR BAR" , fontsize = 15)

plt.xlabel("horizontal" , fontsize = 15 , color = "blue")
plt.ylabel("vertical" , fontsize = 15  , color = "blue")

plt.title("SCATTER" , fontsize = 25 , color = "blue")
plt.show()
```

```python
import matplotlib.pyplot as plt
list = [1 , 2 , 3 , 4 , 5 , 6 , 7 , 8]
list_A = [0 , 1 , 2 , 6 ,  3 , 9, 8 , 3]
colors = [10, 40, 50, 45, 90, 30, 20, 65]

sizes = [100 , 290 , 500 , 110 , 200 , 300 , 600 , 120]
plt.scatter(list , list_A , c = colors , cmap="Dark2" , s = sizes)

t = plt.colorbar()
t.set_label("COLOR BAR" , fontsize = 15 , color = 'g')

plt.xlabel("horizontally" , fontsize = 15 , color = "blue")
plt.ylabel("vertically" , fontsize = 15  , color = "blue")
plt.xticks(list , rotation = 30)
plt.title("SCATTER" , fontsize = 25 , color = "blue")
plt.show()
```
